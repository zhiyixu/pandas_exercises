```python
tips = pd.read_csv("E:/#/tips.csv", sep=",")
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop("Unnamed: 0", inplace=True, axis=1)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
fig, ax = plt.subplots(1,1)
ax.hist(tips.total_bill)
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
fig, ax = plt.subplots(1,1)
ax.scatter(tips.total_bill, tips.tip)
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
tips.head()
fig, ax = plt.subplots(1,1,figsize=(10, 10))

ax.scatter(tips.day, tips.tip)
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
fig, ax = plt.subplots(1,1,figsize=(10, 5))

tips["c"] = tips["sex"].apply(lambda x: 0 if x=='Male' else 1)

ax.scatter(tips.tip, tips.day, c=tips.c)
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
tips.head()
fig, ax = plt.subplots(1,2,figsize=(10, 5), sharey=True)

d = tips.query("time=='Dinner'").copy()
l = tips.query("time=='Lunch'").copy()

ax[0].boxplot(d.tip)
ax[1].boxplot(l.tip)

ax[0].set_ylabel("Tip")
ax[0].set_xlabel("Dinner")
ax[1].set_xlabel("Lunch")
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tips.head()
fig, ax = plt.subplots(1,2,figsize=(10, 5), sharey=True)

d = tips.query("sex=='Male'").copy()
l = tips.query("sex=='Female'").copy()

ax[0].scatter(d.tip, d.total_bill, c=d.c, label="Somker")
ax[1].scatter(l.tip, l.total_bill, c=l.c, label="Somker")

ax[0].set_ylabel("Tip")
ax[0].set_title("Male")
ax[0].set_xlabel("Total Bill")
ax[0].legend(loc='best')
ax[1].set_xlabel("Total Bill")
ax[1].set_title("Female")
ax[1].legend(loc='best')

plt.show()
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)

g.add_legend();
```
