### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("../../#/chipotle.tsv", sep="\t")
chipo.head(10)
```

### Step 4. How many products cost more than $10.00?

```python
func_price = lambda x: round(float(x.replace("$","")))
chipo["price"] = chipo["item_price"].apply(func_price)
chipo[chipo["price"] > 10]["item_name"].value_counts().count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
df = chipo[chipo["price"] > 10]
df = df[["item_name","item_price"]].drop_duplicates()
df
```

### Step 6. Sort by the name of the item

```python
df.sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.head()
chipo.groupby(by=["price"]).sum("quantity").sort_values("quantity", ascending=False).iloc[0]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo["item_name"]=="Veggie Salad Bowl"].count().order_id
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
soda = chipo[chipo["item_name"]=="Veggie Salad Bowl"].count().order_id
others = chipo.groupby(by=["item_name"]).count()
others["frec"] = others["order_id"] - soda
others[others["frec"]>0]
```
